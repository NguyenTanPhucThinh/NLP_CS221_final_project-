```python
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import pandas as pd


def visualize_data_distribution(file_path=None, text_col="text", max_classes_to_plot=20):
    """
    Visualize class balance and text-length behavior for NLP datasets.

    Works with train.csv that uses columns like `act` and `emotion`, and
    also supports datasets with a single `label` column.
    """
    sns.set_theme(style="whitegrid")

    # Resolve dataset path
    if file_path is None:
        file_path = Path("/content/drive/MyDrive/NLP_final_project/NLP_main/data/train.csv") # Convert default string to Path object
    else:
        file_path = Path(file_path)

    if not file_path.exists():
        raise FileNotFoundError(f"Dataset not found: {file_path}")

    # Load data
    df = pd.read_csv(file_path)

    if text_col not in df.columns:
        raise ValueError(
            f"Expected text column '{text_col}' not found. Available columns: {list(df.columns)}"
        )

    # Text-derived features
    df[text_col] = df[text_col].astype(str)
    df["text_length"] = df[text_col].str.len()
    df["word_count"] = df[text_col].str.split().map(len)

    # Candidate target columns: non-text columns with manageable cardinality
    excluded = {text_col, "text_length", "word_count"}
    candidate_targets = []
    for col in df.columns:
        if col in excluded:
            continue
        nunique = df[col].nunique(dropna=True)
        if nunique > 1 and nunique <= max_classes_to_plot:
            candidate_targets.append(col)

    print("=== Dataset Overview ===")
    print(f"Path: {file_path}")
    print(f"Total samples: {len(df):,}")
    print(f"Columns: {list(df.columns)}")
    print()

    print("=== Missing Values ===")
    print(df.isna().sum())
    print()

    print("=== Text Statistics ===")
    text_stats = df[["text_length", "word_count"]].describe(percentiles=[0.5, 0.9, 0.95])
    print(text_stats)
    print()

    # Global distributions for text size
    fig, axes = plt.subplots(1, 2, figsize=(16, 5))
    sns.histplot(df["text_length"], bins=50, kde=True, color="#4C78A8", ax=axes[0])
    axes[0].set_title("Text Length Distribution")
    axes[0].set_xlabel("Characters")

    sns.histplot(df["word_count"], bins=50, kde=True, color="#F58518", ax=axes[1])
    axes[1].set_title("Word Count Distribution")
    axes[1].set_xlabel("Words")

    plt.tight_layout()
    plt.show()

    if not candidate_targets:
        print("No categorical target columns detected for label-wise visualizations.")
        return

    print("=== Target Column Summary ===")
    print(f"Detected target columns: {candidate_targets}")
    print()

    # Per-target visualizations and stats
    for target_col in candidate_targets:
        print(f"--- {target_col} distribution ---")
        counts = df[target_col].value_counts(dropna=False)
        percentages = (counts / len(df) * 100).round(2)
        summary = pd.DataFrame({"count": counts, "percent": percentages})
        print(summary)
        print()

        top_categories = counts.head(max_classes_to_plot).index
        plot_df = df[df[target_col].isin(top_categories)].copy()

        fig, axes = plt.subplots(1, 2, figsize=(18, 5))

        # Class balance
        sns.countplot(
            data=plot_df,
            x=target_col,
            order=counts.head(max_classes_to_plot).index,
            palette="Set2",
            ax=axes[0],
        )
        axes[0].set_title(f"Class Distribution: {target_col}")
        axes[0].set_xlabel(target_col)
        axes[0].set_ylabel("Count")
        axes[0].tick_params(axis="x", rotation=45)

        # Length by class
        sns.boxplot(
            data=plot_df,
            x=target_col,
            y="text_length",
            order=counts.head(max_classes_to_plot).index,
            palette="Set3",
            ax=axes[1],
        )
        axes[1].set_title(f"Text Length by {target_col}")
        axes[1].set_xlabel(target_col)
        axes[1].set_ylabel("Characters")
        axes[1].tick_params(axis="x", rotation=45)

        plt.tight_layout()
        plt.show()

        # Per-class stats
        class_stats = (
            df.groupby(target_col)[["text_length", "word_count"]]
            .describe(percentiles=[0.5, 0.9, 0.95])
            .round(2)
        )
        print(f"--- {target_col} text stats ---")
        print(class_stats)
        print()

    # If act and emotion both exist, show relationship heatmap
    if {"act", "emotion"}.issubset(df.columns):
        cross = pd.crosstab(df["act"], df["emotion"])
        plt.figure(figsize=(10, 7))
        sns.heatmap(cross, cmap="YlGnBu", annot=False, cbar=True)
        plt.title("Act vs Emotion (Count Heatmap)")
        plt.xlabel("emotion")
        plt.ylabel("act")
        plt.tight_layout()
        plt.show()



# Call the function
visualize_data_distribution()
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
from transformers import BertModel, BertTokenizer
import numpy as np
import pandas as pd # Ensure pandas is imported for pd.read_csv

# Create label mappings for 'act' and 'emotion' columns
unique_act_labels_raw = sorted(train_df['act'].unique().tolist())
act_label_to_id = {label_val: i for i, label_val in enumerate(unique_act_labels_raw)}
id_to_act_label = {i: label_val for i, label_val in enumerate(unique_act_labels_raw)}

unique_emotion_labels_raw = sorted(train_df['emotion'].unique().tolist())
emotion_label_to_id = {label_val: i for i, label_val in enumerate(unique_emotion_labels_raw)}
id_to_emotion_label = {i: label_val for i, label_val in enumerate(unique_emotion_labels_raw)}

print(f"\nUnique 'act' labels extracted: {unique_act_labels_raw[:5]}...")
print(f"Sample act_label_to_id mapping: {{0: id_to_act_label.get(0, 'N/A'), 1: id_to_act_label.get(1, 'N/A')}}...")
print(f"\nUnique 'emotion' labels extracted: {unique_emotion_labels_raw[:5]}...")
print(f"Sample emotion_label_to_id mapping: {{0: id_to_emotion_label.get(0, 'N/A'), 1: id_to_emotion_label.get(1, 'N/A')}}...")

# Convert string labels to integer IDs in all dataframes
train_df['act_label_id'] = train_df['act'].map(act_label_to_id)
train_df['emotion_label_id'] = train_df['emotion'].map(emotion_label_to_id)

val_df['act_label_id'] = val_df['act'].map(act_label_to_id)
val_df['emotion_label_id'] = val_df['emotion'].map(emotion_label_to_id)

test_df['act_label_id'] = test_df['act'].map(act_label_to_id)
test_df['emotion_label_id'] = test_df['emotion'].map(emotion_label_to_id)


class MultiTaskTextDataset(Dataset):
    def __init__(self, dataframe, tokenizer, max_length=128):
        self.dataframe = dataframe
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.dataframe)

    def __getitem__(self, idx):
        text = str(self.dataframe.iloc[idx]['text'])
        act_label = int(self.dataframe.iloc[idx]['act_label_id'])
        emotion_label = int(self.dataframe.iloc[idx]['emotion_label_id'])

        # Tokenize the text
        encoding = self.tokenizer(
            text,
            truncation=True,
            padding='max_length',
            max_length=self.max_length,
            return_tensors='pt'
        )

        return {
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
            'act_label': torch.tensor(act_label, dtype=torch.long),
            'emotion_label': torch.tensor(emotion_label, dtype=torch.long)
        }

# Create datasets with max_length=64 (as determined from previous run)
train_dataset = MultiTaskTextDataset(train_df, tokenizer, max_length=64)
val_dataset = MultiTaskTextDataset(val_df, tokenizer, max_length=64)
test_dataset = MultiTaskTextDataset(test_df, tokenizer, max_length=64)

# Create dataloaders
batch_size = 16
train_dataloader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_dataloader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
test_dataloader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

print("DataLoaders created successfully for multi-task!")
print(f"Max sequence length: 64")
print(f"Train batches: {len(train_dataloader)}")
print(f"Val batches: {len(val_dataloader)}")
print(f"Test batches: {len(test_dataloader)}")
```

```python
import torch
from sklearn.utils.class_weight import compute_class_weight
import numpy as np

# Setup device early so weights can be moved to it
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device for weights: {device}")

# Emotion weights
# Prepare labels for compute_class_weight. Use the actual mapped IDs from train_df.
emotion_labels_for_weights = train_df['emotion_label_id'].tolist()
emotion_class_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.arange(len(unique_emotion_labels_raw)), # All possible class IDs
    y=emotion_labels_for_weights
)
emotion_weights = torch.tensor(emotion_class_weights, dtype=torch.float32).to(device)

# Act weights
# Prepare labels for compute_class_weight. Use the actual mapped IDs from train_df.
act_labels_for_weights = train_df['act_label_id'].tolist()
act_class_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.arange(len(unique_act_labels_raw)), # All possible class IDs
    y=act_labels_for_weights
)
act_weights = torch.tensor(act_class_weights, dtype=torch.float32).to(device)

print("Class weights calculated and moved to device:")
print("Act weights:", act_weights)
print("Emotion weights:", emotion_weights)
```

Now that we have the class weights, we'll modify the `MultiTaskBert` model to use these weights in its `CrossEntropyLoss` functions. We'll also update the training and evaluation functions to remove the `loss_fn` parameter, as loss calculation will now be handled internally by the model.

```python
from transformers import BertModel, BertConfig
from torch import nn

class MultiTaskBert(nn.Module):
    def __init__(self, config: BertConfig):
        super().__init__()
        self.config = config
        self.bert = BertModel(config)
        self.dropout = nn.Dropout(config.hidden_dropout_prob)

        self.act_classifier = nn.Linear(config.hidden_size, config.num_labels_act)
        self.emotion_classifier = nn.Linear(config.hidden_size, config.num_labels_emotion)

    def forward(
        self,
        input_ids=None,
        attention_mask=None,
        token_type_ids=None,
        position_ids=None,
        head_mask=None,
        inputs_embeds=None,
        act_labels=None,
        emotion_labels=None,
        output_attentions=None,
        output_hidden_states=None,
        return_dict=None,
    ):
        return_dict_self = return_dict if return_dict is not None else self.config.use_return_dict

        outputs = self.bert(
            input_ids,
            attention_mask=attention_mask,
            token_type_ids=token_type_ids,
            position_ids=position_ids,
            head_mask=head_mask,
            inputs_embeds=inputs_embeds,
            output_attentions=output_attentions,
            output_hidden_states=output_hidden_states,
            return_dict=True,
        )

        pooled_output = outputs.pooler_output
        pooled_output = self.dropout(pooled_output)

        act_logits = self.act_classifier(pooled_output)
        emotion_logits = self.emotion_classifier(pooled_output)

        loss = None
        if act_labels is not None and emotion_labels is not None:
            # Use class weights in CrossEntropyLoss
            act_loss_fct = nn.CrossEntropyLoss(weight=act_weights)
            emotion_loss_fct = nn.CrossEntropyLoss(weight=emotion_weights)
            act_loss = act_loss_fct(act_logits.view(-1, self.config.num_labels_act), act_labels.view(-1))
            emotion_loss = emotion_loss_fct(emotion_logits.view(-1, self.config.num_labels_emotion), emotion_labels.view(-1))
            loss = act_loss + emotion_loss

        if not return_dict_self:
            output = (act_logits, emotion_logits)
            if output_hidden_states:
                output = output + (outputs.hidden_states,)
            if output_attentions:
                output = output + (outputs.attentions,)
            return ((loss,) + output) if loss is not None else output

        res_dict = {
            "loss": loss,
            "act_logits": act_logits,
            "emotion_logits": emotion_logits,
        }
        if output_hidden_states:
            res_dict["hidden_states"] = outputs.hidden_states
        if output_attentions:
            res_dict["attentions"] = outputs.attentions
        return res_dict

print("MultiTaskBert model class defined.")
```

```python
from typing import Dict, List, Tuple, Optional # Added Optional
from tqdm.auto import tqdm
import torch.nn.functional as F

def train_step(
    model: torch.nn.Module,
    dataloader: torch.utils.data.DataLoader,
    # loss_fn: torch.nn.Module, # Removed as loss is calculated internally by the model
    optimizer: torch.optim.Optimizer,
    device: torch.device
) -> Tuple[float, float, float]: # Return loss, act_acc, emotion_acc
    """Single training epoch.

    Args:
        model: PyTorch model to train.
        dataloader: Training DataLoader.
        # loss_fn: Loss function. # Removed
        optimizer: Optimizer.
        device: Device to compute on.

    Returns:
        Tuple of (average_loss, act_accuracy, emotion_accuracy)
    """
    model.train()
    total_loss = 0.0
    total_act_correct = 0
    total_emotion_correct = 0
    total_samples = 0

    for batch in tqdm(dataloader, desc="Training", leave=False):
        # Move batch to device
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        act_labels = batch['act_label'].to(device)  # Corrected key
        emotion_labels = batch['emotion_label'].to(device) # Corrected key

        # Forward pass
        outputs = model(input_ids=input_ids,
                       attention_mask=attention_mask,
                       act_labels=act_labels,
                       emotion_labels=emotion_labels)
        loss = outputs['loss'] # Corrected: Access loss as a dictionary key

        # Backward pass
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        # Track metrics
        total_loss += loss.item() * input_ids.size(0)

        act_predictions = torch.argmax(outputs['act_logits'], dim=1)
        total_act_correct += (act_predictions == act_labels).sum().item()

        emotion_predictions = torch.argmax(outputs['emotion_logits'], dim=1)
        total_emotion_correct += (emotion_predictions == emotion_labels).sum().item()

        total_samples += input_ids.size(0) # Use input_ids.size(0) as batch size

    avg_loss = total_loss / total_samples
    act_accuracy = total_act_correct / total_samples
    emotion_accuracy = total_emotion_correct / total_samples

    return avg_loss, act_accuracy, emotion_accuracy


def test_step(
    model: torch.nn.Module,
    dataloader: torch.utils.data.DataLoader,
    # loss_fn: torch.nn.Module, # Removed as loss is calculated internally by the model
    device: torch.device
) -> Tuple[float, float, float]: # Return loss, act_acc, emotion_acc
    """Single evaluation epoch.

    Args:
        model: PyTorch model to evaluate.
        dataloader: Evaluation DataLoader.
        # loss_fn: Loss function. # Removed
        device: Device to compute on.

    Returns:
        Tuple of (average_loss, act_accuracy, emotion_accuracy)
    """
    model.eval()
    total_loss = 0.0
    total_act_correct = 0
    total_emotion_correct = 0
    total_samples = 0

    with torch.no_grad():
        for batch in tqdm(dataloader, desc="Evaluating", leave=False):
            # Move batch to device
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            act_labels = batch['act_label'].to(device) # Corrected key
            emotion_labels = batch['emotion_label'].to(device) # Corrected key

            # Forward pass
            outputs = model(input_ids=input_ids,
                           attention_mask=attention_mask,
                           act_labels=act_labels,
                           emotion_labels=emotion_labels)
            loss = outputs['loss'] # Corrected: Access loss as a dictionary key

            # Track metrics
            total_loss += loss.item() * input_ids.size(0)

            act_predictions = torch.argmax(outputs['act_logits'], dim=1)
            total_act_correct += (act_predictions == act_labels).sum().item()

            emotion_predictions = torch.argmax(outputs['emotion_logits'], dim=1)
            total_emotion_correct += (emotion_predictions == emotion_labels).sum().item()

            total_samples += input_ids.size(0)

    avg_loss = total_loss / total_samples
    act_accuracy = total_act_correct / total_samples
    emotion_accuracy = total_emotion_correct / total_samples

    return avg_loss, act_accuracy, emotion_accuracy


def train(
    model: torch.nn.Module,
    train_dataloader: torch.utils.data.DataLoader,
    val_dataloader: torch.utils.data.DataLoader,
    optimizer: torch.optim.Optimizer,
    # loss_fn: torch.nn.Module, # Removed
    epochs: int,
    device: torch.device,
    early_stopping_patience: int = 3,
    lr_scheduler: Optional[torch.optim.lr_scheduler._LRScheduler] = None
) -> Dict[str, List[float]]:
    """Trains and evaluates a BERT model with early stopping.

    Args:
        model: PyTorch model to train.
        train_dataloader: Training DataLoader.
        val_dataloader: Validation DataLoader.
        optimizer: Optimizer.
        # loss_fn: Loss function. # Removed
        epochs: Number of epochs to train.
        device: Device to compute on.
        early_stopping_patience: Epochs to wait before early stopping.
        lr_scheduler: Optional learning rate scheduler.

    Returns:
        Dictionary with training/validation metrics for each epoch.
    """
    results = {
        "train_loss": [],
        "train_act_acc": [],
        "train_emotion_acc": [],
        "val_loss": [],
        "val_act_acc": [],
        "val_emotion_acc": []
    }

    best_val_loss = float('inf')
    patience_counter = 0

    for epoch in tqdm(range(epochs), desc="Epochs"):
        # Training phase
        train_loss, train_act_acc, train_emotion_acc = train_step(
            model=model,
            dataloader=train_dataloader,
            # loss_fn=loss_fn, # Removed
            optimizer=optimizer,
            device=device
        )

        # Validation phase
        val_loss, val_act_acc, val_emotion_acc = test_step(
            model=model,
            dataloader=val_dataloader,
            # loss_fn=loss_fn, # Removed
            device=device
        )

        # Step the learning rate scheduler if provided
        if lr_scheduler is not None:
            lr_scheduler.step(val_loss)
            print(f"  Learning Rate: {optimizer.param_groups[0]['lr']:.6f}") # Print current LR

        # Print metrics
        print(
            f"Epoch {epoch+1}/{epochs} | "
            f"Train Loss: {train_loss:.4f} | Train Act Acc: {train_act_acc:.4f} | Train Emotion Acc: {train_emotion_acc:.4f} | "
            f"Val Loss: {val_loss:.4f} | Val Act Acc: {val_act_acc:.4f} | Val Emotion Acc: {val_emotion_acc:.4f}"
        )

        # Store results
        results["train_loss"].append(train_loss)
        results["train_act_acc"].append(train_act_acc)
        results["train_emotion_acc"].append(train_emotion_acc)
        results["val_loss"].append(val_loss)
        results["val_act_acc"].append(val_act_acc)
        results["val_emotion_acc"].append(val_emotion_acc)

        # Early stopping
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            patience_counter = 0
            # Save best model
            torch.save(model.state_dict(), 'best_model.pt')
        else:
            patience_counter += 1
            if patience_counter >= early_stopping_patience:
                print(f"\nEarly stopping triggered after {epoch+1} epochs")
                # Load best model, explicitly mapping to device
                model.load_state_dict(torch.load('best_model.pt', map_location=device))
                break

    return results
```

```python
# Setup training configuration
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")
# Move model to device
multi_task_model = multi_task_model.to(device)

# Define optimizer and loss function
optimizer = torch.optim.AdamW(multi_task_model.parameters(), lr=2e-5)
# loss_fn = torch.nn.CrossEntropyLoss() # Removed as loss is calculated internally by the model

# Define learning rate scheduler (ReduceLROnPlateau)
# It reduces the learning rate when a metric (val_loss) has stopped improving.
from torch.optim.lr_scheduler import ReduceLROnPlateau
scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.1, patience=1)

# Train the model
epochs = 5
results = train(
    model=multi_task_model,
    train_dataloader=train_dataloader,
    val_dataloader=val_dataloader,
    optimizer=optimizer,
    epochs=epochs,
    device=device,
    early_stopping_patience=3,
    lr_scheduler=scheduler # Pass the scheduler to the train function
)

print("\nTraining complete!")
# print(f"Best validation act accuracy: {max(results['val_act_acc']):.4f}")
# print(f"Final test act accuracy: {results['test_act_acc'][-1]:.4f}")
# print(f"Best validation emotion accuracy: {max(results['val_emotion_acc']):.4f}")
# print(f"Final test emotion accuracy: {results['test_emotion_acc'][-1]:.4f}")
```

```python
import torch
import pandas as pd
from tqdm.auto import tqdm
from transformers import AutoTokenizer
from torch.utils.data import Dataset, DataLoader

# Ensure test_df is loaded (redundant if run after kJ19TskzJVOj, but ensures self-containment)
test_csv_path = "/content/drive/MyDrive/NLP_final_project/NLP_main/data/test.csv"
test_df = pd.read_csv(test_csv_path)

# Ensure tokenizer is loaded (redundant if run after 7vhQQfk9Jema, but ensures self-containment)
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

# Ensure MultiTaskTextDataset is defined and modified to return text
# This class definition is copied from qf9ON4dBKFae but modified.
class MultiTaskTextDataset(Dataset):
    def __init__(self, dataframe, tokenizer, max_length=128):
        self.dataframe = dataframe
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.dataframe)

    def __getitem__(self, idx):
        text = str(self.dataframe.iloc[idx]['text'])
        # Ensure act_label_id and emotion_label_id exist, using 0 as a default if not found
        # These mappings (act_label_to_id, emotion_label_to_id) are expected to be available
        # from the execution of cell 712213c4 or qf9ON4dBKFae.
        act_label = int(self.dataframe.iloc[idx]['act_label_id']) if 'act_label_id' in self.dataframe.columns else 0
        emotion_label = int(self.dataframe.iloc[idx]['emotion_label_id']) if 'emotion_label_id' in self.dataframe.columns else 0

        # Tokenize the text
        encoding = self.tokenizer(
            text,
            truncation=True,
            padding='max_length',
            max_length=self.max_length,
            return_tensors='pt'
        )

        return {
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
            'act_label': torch.tensor(act_label, dtype=torch.long),
            'emotion_label': torch.tensor(emotion_label, dtype=torch.long),
            'text': text # IMPORTANT: Return the original text
        }

# Create label IDs for test_df if they don't exist, as MultiTaskTextDataset expects them
# These mappings (act_label_to_id, emotion_label_to_id) are expected to be available from cell 712213c4
if 'act_label_id' not in test_df.columns:
    # Assuming act_label_to_id is globally available from a previous cell (e.g., 712213c4 or qf9ON4dBKFae)
    test_df['act_label_id'] = test_df['act'].map(act_label_to_id)
if 'emotion_label_id' not in test_df.columns:
    # Assuming emotion_label_to_id is globally available from a previous cell
    test_df['emotion_label_id'] = test_df['emotion'].map(emotion_label_to_id)

# Ensure test_dataloader is created or re-created with the updated dataset class
batch_size = 16 # Assuming a common batch size, match qf9ON4dBKFae
max_length = 64 # Assuming a common max_length, match qf9ON4dBKFae
test_dataset = MultiTaskTextDataset(test_df, tokenizer, max_length=max_length)
test_dataloader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

# Collect predictions and true labels for the test set
all_texts = []
all_act_labels = []
all_act_preds = []
all_emotion_labels = []
all_emotion_preds = []

with torch.no_grad():
    for batch in tqdm(test_dataloader, desc="Collecting predictions", leave=False):
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        act_labels = batch['act_label']
        emotion_labels = batch['emotion_label']
        current_batch_texts = batch['text'] # Directly get original text from the batch

        outputs = multi_task_model_loaded(input_ids=input_ids, attention_mask=attention_mask)
        act_predictions = torch.argmax(outputs['act_logits'], dim=1)
        emotion_predictions = torch.argmax(outputs['emotion_logits'], dim=1)

        all_texts.extend(current_batch_texts)
        all_act_labels.extend(act_labels.cpu().numpy())
        all_act_preds.extend(act_predictions.cpu().numpy())
        all_emotion_labels.extend(emotion_labels.cpu().numpy())
        all_emotion_preds.extend(emotion_predictions.cpu().numpy())
```
